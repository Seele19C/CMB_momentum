```python
#compensate
def momentum_00_compensate(cat, Nmesh):
    if 'weight' not in cat:
        #无权重
        cat['weight'] = 1
    if 'RsdPos_x' in cat:
        #有红移
        LOS=(1,0,0)
        cat['RsdPos_x'] = cat['RsdPos_x']%1000
        # momentum_mesh = cat.to_mesh(interlaced=True,compensated=False, Nmesh=256, BoxSize=1000, window='tsc', position='RsdPos_x', value='Vx',weight='weight')
        mesh = cat.to_mesh(interlaced=True,compensated=True, Nmesh=Nmesh, BoxSize=1000, window='tsc', position='RsdPos_x')
    else:
        #无红移
        # momentum_mesh = cat.to_mesh(interlaced=False,compensated=False, Nmesh=256, BoxSize=1000, window='tsc', position='Pos', value='Vx',weight='weight')
        mesh = cat.to_mesh(interlaced=True,compensated=True, Nmesh=Nmesh, BoxSize=1000, window='tsc', position='Pos')
    r00 = FFTPower(mesh, mode='1d', los=[1,0,0])
    P00 = r00.power['power'].real - r00.attrs['shotnoise']
    #cache
    # if cache is not False: 
    #     cache_write(num=num, r01=r01, P01=P01, gamma=gamma)
    return r00, P00

#no compensate
def momentum_00_nocompensate(cat, Nmesh):
    if 'weight' not in cat:
        #无权重
        cat['weight'] = 1
    if 'RsdPos_x' in cat:
        #有红移
        LOS=(1,0,0)
        cat['RsdPos_x'] = cat['RsdPos_x']%1000
        # momentum_mesh = cat.to_mesh(interlaced=True,compensated=False, Nmesh=256, BoxSize=1000, window='tsc', position='RsdPos_x', value='Vx',weight='weight')
        mesh = cat.to_mesh(interlaced=True,compensated=False, Nmesh=Nmesh, BoxSize=1000, window='tsc', position='RsdPos_x')
    else:
        #无红移
        # momentum_mesh = cat.to_mesh(interlaced=False,compensated=False, Nmesh=256, BoxSize=1000, window='tsc', position='Pos', value='Vx',weight='weight')
        mesh = cat.to_mesh(interlaced=True,compensated=False, Nmesh=Nmesh, BoxSize=1000, window='tsc', position='Pos')
    r00 = FFTPower(mesh, mode='1d', los=[1,0,0])
    P00 = r00.power['power'].real - r00.attrs['shotnoise']
    #cache
    # if cache is not False: 
    #     cache_write(num=num, r01=r01, P01=P01, gamma=gamma)
    return r00, P00

def momentum_00_nointerlaced(cat, Nmesh):
    if 'weight' not in cat:
        #无权重
        cat['weight'] = 1
    if 'RsdPos_x' in cat:
        #有红移
        LOS=(1,0,0)
        cat['RsdPos_x'] = cat['RsdPos_x']%1000
        # momentum_mesh = cat.to_mesh(interlaced=True,compensated=False, Nmesh=256, BoxSize=1000, window='tsc', position='RsdPos_x', value='Vx',weight='weight')
        mesh = cat.to_mesh(interlaced=False,compensated=False, Nmesh=Nmesh, BoxSize=1000, window='tsc', position='RsdPos_x')
    else:
        #无红移
        # momentum_mesh = cat.to_mesh(interlaced=False,compensated=False, Nmesh=256, BoxSize=1000, window='tsc', position='Pos', value='Vx',weight='weight')
        mesh = cat.to_mesh(interlaced=False,compensated=False, Nmesh=Nmesh, BoxSize=1000, window='tsc', position='Pos')
    r00 = FFTPower(mesh, mode='1d', los=[1,0,0])
    P00 = r00.power['power'].real - r00.attrs['shotnoise']
    #cache
    # if cache is not False: 
    #     cache_write(num=num, r01=r01, P01=P01, gamma=gamma)
    return r00, P00
```

```python
import numpy as np
num_test = 12
cat = m.data_read(num_test)
cat_nw = m.RED(cat)
del cat
for num in [256, 512, 1024]:
    r01_nw, P01_nw = momentum_00_compensate(cat_nw, num)
    r01_nw_nc, P01_nw_nc = momentum_00_nocompensate(cat_nw, num)
    r01_nw_ni, P01_nw_ni = momentum_00_nointerlaced(cat_nw, num)

    if num == 256 : 
        k256 = r01_nw.power['k']
        del r01_nw
        P256 = P01_nw
        del P01_nw
        k256_nc = r01_nw_nc.power['k']
        del r01_nw_nc
        P256_nc = P01_nw_nc
        del P01_nw_nc
        k256_ni = r01_nw_ni.power['k']
        del r01_nw_ni
        P256_ni = P01_nw_ni
        del P01_nw_ni
        gc.collect()
    elif num == 512 : 
        k512 = r01_nw.power['k']
        del r01_nw
        P512 = P01_nw
        del P01_nw
        k512_nc = r01_nw_nc.power['k']
        del r01_nw_nc
        P512_nc = P01_nw_nc
        del P01_nw_nc
        k512_ni = r01_nw_ni.power['k']
        del r01_nw_ni
        P512_ni = P01_nw_ni
        del P01_nw_ni
        gc.collect()
    elif num == 1024 :
        k1024 = r01_nw.power['k']
        del r01_nw
        P1024 = P01_nw  
        del P01_nw
        k1024_nc = r01_nw_nc.power['k']
        del r01_nw_nc
        P1024_nc = P01_nw_nc
        del P01_nw_nc
        k1024_ni = r01_nw_ni.power['k']
        del r01_nw_ni
        P1024_ni = P01_nw_ni
        del P01_nw_ni
        gc.collect()
   


lines = [
    ('256',     k256,     P256),
    ('256_nc',     k256_nc,     P256_nc),
    ('256_ni',     k256_ni,     P256_ni),
    ('512',     k512,     P512),
    ('512_nc',     k512_nc,     P512_nc),
    ('512_ni',     k512_ni,     P512_ni),
    ('1024',     k1024,     P1024),
    ('1024_nc',     k1024_nc,     P1024_nc),
    ('1024_ni',     k1024_ni,     P1024_ni),

]


target_k = 0.01

_, k_ref, P_ref = lines[0]
idx_ref = np.argmin(np.abs(k_ref - target_k))
P_ref_value = P_ref[idx_ref]

plt.figure()
for name, k_arr, P_arr in lines:
    idx = np.argmin(np.abs(k_arr - target_k))
    factor = P_ref_value / P_arr[idx]
    P_aligned = P_arr * factor
    # ls = '--' if 'nc' in name else ':' if 'ni' in name else '-.'
    marker = '|' if '256' in name else 'x' if '512' in name else '+'
    plt.loglog(k_arr, P_aligned, linestyle='-', label=name, marker = marker)

# for i, (k_i, P_i) in enumerate(lines):
#     idx_i = np.argmin(np.abs(k_i - target_k))
#     P_i_value = P_i[idx_i]
#     factor = P_ref_value / P_i_value
#     P_i_aligned = P_i * factor
#     plt.loglog(k_i, P_i_aligned, label=f'Line {i+1} aligned')




plt.xlabel('k')
plt.ylabel('P(k)')
plt.title('Align curves at k=0.01')
plt.legend()
plt.show()
```
